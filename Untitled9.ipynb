```python
# Import required libraries
import torch
from diffusers import StableDiffusionPipeline
import numpy as np
import cv2
from PIL import Image
from IPython.display import display, Video

class TextToVideoGenerator:
    def __init__(self, model_name="stabilityai/stable-diffusion-2-1-base", device="cpu"):
        """Initialize the text-to-video pipeline."""
        print(f"Loading model: {model_name}")
        self.pipe = StableDiffusionPipeline.from_pretrained(model_name)
        self.pipe.to(device)
        self.device = device
        print("Model loaded successfully on CPU.")

    def generate_frames(self, prompt, num_frames=3, image_size=(256, 256)):
        """Generate a small series of frames from the input text."""
        print(f"Generating frames for prompt: '{prompt}'")
        frames = []
        for i in range(num_frames):
            print(f"Generating frame {i + 1}/{num_frames}")
            result = self.pipe(prompt, height=image_size[0], width=image_size[1])
            frame = result.images[0]
            frames.append(np.array(frame))  # Convert PIL Image to NumPy array
        return frames

    def save_video(self, frames, fps=2, output_path="optimized_video.mp4"):
        """Save a series of frames as a video."""
        print("Saving frames as a video...")
        height, width, _ = frames[0].shape
        fourcc = cv2.VideoWriter_fourcc(*"mp4v")
        video_writer = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        for frame in frames:
            video_writer.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))  # Convert RGB to BGR

        video_writer.release()
        print(f"Video saved at {output_path}")
        return output_path

# Main execution
if __name__ == "__main__":
    # Step 1: Ask user for input parameters
    text_prompt = input("Enter the text prompt for video generation: ")
    num_frames = int(input("Enter the number of frames to generate (e.g., 3): "))
    fps = int(input("Enter the frames per second (fps) for the video: "))
    image_size = tuple(map(int, input("Enter image size (width, height) as comma-separated values (e.g., 256,256): ").split(',')))

    # Step 2: Generate and save the video
    generator = TextToVideoGenerator(device="cpu")
    frames = generator.generate_frames(text_prompt, num_frames=num_frames, image_size=image_size)
    output_path = generator.save_video(frames, fps=fps)

    # Step 3: Display the generated video
    display(Video(output_path, embed=True))
```
